```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def bloom(prob, no_dumps):
    return 1 - (1 - prob)**no_dumps
```

```python
fprates = np.linspace(0.01, 0.02, 5)
x = np.linspace(1,10,10)
fig = plt.figure()
for fp in fprates:
    plt.plot(x, [bloom(fp, el) for el in x], label=fp)
plt.xlabel('number of dumps', fontsize=18)
plt.ylabel('Collective FP rate', fontsize=16)
plt.legend(loc='upper left')
```
